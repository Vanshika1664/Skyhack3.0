# **Flight Difficulty Score Development**
Build a systematic daily-level scoring approach that resets every day. Below are the outputs required:

Ranking: Within each day, order flights by their difficulty score so that the highest-ranked flights represent the most difficult to manage. Classification: Group flights into three categories  Difficult, Medium, or Easy  based on rank distribution.

This code calculates four stress metrics for each flight: ground time stress, bag handling stress (transfer vs. origin), passenger load stress, and special service request stress. These metrics are designed to quantify potential operational challenges for each flight. Any missing or infinite values in these stress metrics are then replaced with 0

```python
from sklearn.preprocessing import MinMaxScaler

#create features that indicate "difficulty"

#ground stress: Handle potential division by zero or infinite values
flights['ground_stress'] = 1 - (flights['scheduled_ground_time_minutes'] / (flights['minimum_turn_minutes'] + 1))

#bag stress: Calculate ratio from original bags dataframe and merge
bag_type_counts = bags.groupby('Flight Key')['bag_type'].value_counts().unstack(fill_value=0)
bag_type_counts['bag_stress'] = bag_type_counts['Transfer'] / (bag_type_counts['Origin'] + 1)
flights = pd.merge(flights, bag_type_counts[['bag_stress']], on='Flight Key', how='left')

#load stress: Handle potential division by zero
flights['load_stress'] = flights['total_passengers'] / (flights['total_seats'] + 1)

#service stress: Use number_of_remarks as approximation, handle potential division by zero
flights['service_stress'] = flights['number_of_remarks'] / (flights['total_passengers'] + 1)

#delay stress: Clip negative values



#replace any remaining NaN or infinite values with 0 after calculations
flights = flights.fillna(0)

display(flights.head())
```

This code scales the previously calculated stress features (ground_stress, bag_stress, load_stress, and service_stress) to a range between 0 and 1 using the MinMaxScaler. This makes sure all stress metrics contribute proportionally to the final difficulty score.

```python
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Define the list of stress features to scale
stress_features = ['ground_stress', 'bag_stress', 'load_stress', 'service_stress']

# Apply the scaler to the selected stress features
flights[stress_features] = scaler.fit_transform(flights[stress_features])

# Display the head of the dataframe to show the scaled features
display(flights.head())
```

This code calculates the composite difficulty_score for each flight by taking a weighted average of the scaled ground, bag, load, and service stress scores. Ground stress has the highest weight (40%).

```python
# Calculate the composite difficulty score
flights['difficulty_score'] = (
    0.40 * flights['ground_stress'] +
    0.20 * flights['bag_stress'] +
    0.25 * flights['load_stress'] +
    0.15 * flights['service_stress']
)

# Display the head of the dataframe to show the new difficulty score
display(flights.head())
```

This code calculates the daily rank of the difficulty score for each flight and then sorts the flights within each day by this daily rank, with the most difficult flights appearing first.

```python
#calculate the daily rank of the difficulty score
flights['daily_rank'] = flights.groupby('scheduled_departure_date_local')['difficulty_score'] \
                               .rank(method='dense', ascending=False)

#order flights within each day by their daily difficulty rank (highest difficulty first)
daily_ordered_flights = flights.sort_values(by=['scheduled_departure_date_local', 'daily_rank'])

#display the first few rows of the sorted dataframe
display(daily_ordered_flights.head())
```

This code calculates the tertile boundaries of the daily rank and then classifies flights into three categories  Difficult, Medium, or Easy  based on these rank distributions.

```python
#calculate the tertile boundaries of the daily rank
rank_tertiles = flights['daily_rank'].quantile([0.333, 0.667])

#define a function to classify flights based on daily rank tertiles
def classify_difficulty(rank):
    if rank <= rank_tertiles[0.333]:
        return 'Difficult'
    elif rank <= rank_tertiles[0.667]:
        return 'Medium'
    else:
        return 'Easy'

#apply the classification function to create a new 'difficulty_category' column
flights['difficulty_category'] = flights['daily_rank'].apply(classify_difficulty)

#display the counts of flights in each category
display(flights['difficulty_category'].value_counts())

#display the head of the dataframe to show the new classification
display(flights.head())
```

# **Post-Analysis & Operational Insights**




## **Summarize which destinations consistently show more difficulty.**

```python
# now we are extracting only the flight date from the scheduled departure datetime column
# so that we can analyse flights day by day
flights['flight_date'] = pd.to_datetime(flights['scheduled_departure_datetime_local']).dt.date

# now we are calculating the average difficulty score per destination per day
# so that we know on each date which airports had higher or lower scores
daily_difficulty = (flights
                    .groupby(['flight_date','scheduled_arrival_station_code'])['difficulty_score']
                    .mean()
                    .reset_index())

# now we are defining a threshold for difficult days
# so that we can classify the top 20% highest scoring days as difficult
threshold = daily_difficulty['difficulty_score'].quantile(0.8)

# now we are marking each destination-day as difficult or not
# so that later we can count how often each destination falls in the difficult group
consistently_difficult = (daily_difficulty
                          .assign(difficult_day=lambda x: x['difficulty_score'] >= threshold)
                          .groupby('scheduled_arrival_station_code')['difficult_day']
                          .mean()
                          .sort_values(ascending=False))

# now we are printing the top 10 destinations that were most often in the difficult group
# so that we can identify airports that are consistently hard to operate flights into
print("Top 10 consistently difficult destinations:")
print(consistently_difficult.head(10))
```

## **What are the common drivers for those flights?**

```python
# now we are calculating the average stress drivers per destination
# so that we can understand whether ground time, load, bags, or service requests
# are the main reason for difficulty at each airport

drivers_by_dest = (flights
                   .groupby('scheduled_arrival_station_code')[['ground_stress','load_stress','bag_stress','service_stress']]
                   .mean()
                   .sort_values('ground_stress', ascending=False))

# now we are printing the top 5 destinations with their driver averages
print("Average driver values for top destinations:")
print(drivers_by_dest.head(5))
```

###  **Common Drivers Behind Difficult Destinations  Insights**

To understand *why* certain destinations are consistently difficult, we calculated the **average stress values** for each of the four drivers (ground, load, bag, service).  

**Top 5 destinations by driver averages:**
- **AUA**  Ground stress (0.95) and Load stress (0.66) are very high. Bags and service requests play a minimal role.  
- **ILM**  Ground stress (0.95) dominates, with moderate Load stress (0.53). Bag and service stresses are low.  
- **EGE**  Ground stress (0.95) is the key driver, supported by Load stress (0.58).  
- **YHZ**  High Ground stress (0.95) and Load stress (0.59), very low bag/service impact.  
- **NAS**  High Ground stress (0.95) and the **highest Load stress** among this group (0.81).  

**Insights:**
- Across all these destinations, **Ground Stress is the dominant driver**, meaning **tight turnaround times** are the primary operational challenge.  
- **Passenger Load Stress** is the secondary contributor, especially at NAS where passenger volume is unusually high.  
- **Bag Stress** and **Service Stress** values are consistently low, suggesting that baggage handling and special requests are not major issues at these airports compared to time and passenger pressure.  

**Conclusion:**
These results show that the **common drivers of flight difficulty are turnaround pressure and high passenger loads**. Operational improvements at these airports should focus on **increasing buffer times** and **deploying more boarding staff**, rather than baggage or special services.

#  Operational Recommendations Based on Findings

Our analysis combined five datasets (Flights, PNR, Bags, Remarks, Airports) and developed a **Flight Difficulty Score (FDS)**.  
From the results, we derive actionable recommendations for **improving operational efficiency**:

---

## 1. Ground Time Stress
- **Finding:** Ground stress was consistently the *highest driver* of difficulty across airports (average ~0.95).  
- **Example:** Destinations like **AUA, EGE, YHZ** had near-zero turnaround buffers, ranking as consistently difficult.  
- **Action:**  
  - Add **buffer time** in scheduling for airports with historically tight ground margins.  
  - **Prioritise gate allocation** for high-FDS flights to avoid congestion.  
  - Use ORD case study: ORD-origin routes (e.g., ORD  ASE with 142 min delays) clearly need **improved turnaround planning**.

---

## 2. Passenger Load Stress
- **Finding:** High passenger loads increase handling complexity, though not always directly correlated with delay.  
- **Example:** **NAS** had the highest load stress (~0.81) among top difficult airports.  
- **Action:**  
  - Assign **extra boarding staff** for flights with high passenger counts or large PNR groups (>10 pax).  
  - Introduce **digital boarding tools** (self-scan gates, group boarding) to accelerate flow.  

---

## 3. Baggage Complexity
- **Finding:** Flights with disproportionate transfer bags faced higher operational pressure.  
- **Example:** Transfer-to-origin bag ratios showed extreme cases (median 2.0, but some >2500).  
- **Action:**  
  - Allocate **dedicated baggage teams** for flights with high transfer volumes.  
  - Deploy **better tracking** for transfer bags to reduce mishandling.  

---

## 4. Service Requests (Special Passengers)
- **Finding:** Flights with remarks (wheelchairs, strollers, special assistance) had slightly higher delays (~2 minutes more).  
- **Example:** PNR analysis showed **41,739 children and 8,742 stroller users**, requiring boarding assistance.  
- **Action:**  
  - Pre-assign **assistance staff** and equipment (wheelchairs, strollers).  
  - Create **dedicated boarding lanes** for families with children.  

---

## 5. Hub and Route Management
- **Finding:** **ORD (Chicago OHare)** dominates the network and drives most delays.  
- **Example:** 9 of the 10 most delayed routes originated at ORD (e.g., ORD  GRU, ORD  CDG).  
- **Action:**  
  - Treat ORD as a **priority intervention hub**: add staffing buffers, stagger departure slots.  
  - Monitor ORD-heavy flights with **daily FDS alerts** for proactive action.  

---

## Conclusion
The Flight Difficulty Score has highlighted that **tight turnarounds and high passenger loads are the dominant stress factors**, with baggage and special services adding secondary pressure.  
By acting on these findings  **increasing ground buffers, assigning staff strategically, and focusing on critical hubs/routes like ORD**  airlines can significantly improve **on-time performance and operational reliability**.
